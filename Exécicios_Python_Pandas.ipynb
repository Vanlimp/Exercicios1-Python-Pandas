2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mode]

```python
import numpy as np

# Agrupar por profissão e calcular média, mediana e moda
estatisticas_sono = df.groupby("Profissão")["Duração do sono"].agg([
    ('Média', 'mean'),
    ('Mediana', 'median'),
    ('Moda', lambda x: x.mode().tolist())  # Moda pode ter mais de um valor
])

# Exibir resultado
estatisticas_sono
```

3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?


```python
# Filtrar apenas as pessoas que trabalham com Engenharia de Software
engenheiros_software = df[df["Profissão"] == "Eng. de Software"]

# Contar o total de engenheiros de software
total_engenheiros = len(engenheiros_software)

# Contar quantos são classificados como "Obesidade"
obesos = len(engenheiros_software[engenheiros_software["Categoria IMC"] == "Obesidade"])

# Calcular a porcentagem
porcentagem_obesos = (obesos / total_engenheiros) * 100

# Exibir o resultado formatado
print(f"Porcentagem de obesos em Engenharia de Software: {porcentagem_obesos:.2f}%")


```

4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)


```python
# Profissões a serem analisadas
profissoes = ["Advogado(a)", "Representante de Vendas"]

# Filtrar os dados dessas profissões
df_filtrado = df[df["Profissão"].isin(profissoes)]

# Calcular a média de horas de sono dessas profissões
media_sono_profissoes = df_filtrado["Duração do sono"].mean()

# Calcular a média geral de horas de sono
media_sono_geral = df["Duração do sono"].mean()

# Exibir resultados
print(f"Média de horas de sono (Advogado(a) e Representante de Vendas): {media_sono_profissoes:.2f} horas")
print(f"Média geral de horas de sono: {media_sono_geral:.2f} horas")

# Comparação
if media_sono_profissoes < media_sono_geral:
    print("Advogar ou ser representante de vendas faz você dormir menos!")
else:
    print("Não há evidências de que essas profissões dormem menos.")
```

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)

```python
# Definir as profissões a serem comparadas
profissoes = ["Enfermeiro(a)", "Médico(a)"]

# Filtrar os dados dessas profissões
df_filtrado = df[df["Profissão"].isin(profissoes)]

# Calcular a média de horas de sono para cada profissão
media_sono = df_filtrado.groupby("Profissão")["Duração do sono"].mean()

# Exibir os resultados
print(media_sono)

# Comparação
if media_sono["Enfermeiro(a)"] < media_sono["Médico(a)"]:
    print("Enfermeiros(as) dormem menos que médicos(as).")
elif media_sono["Enfermeiro(a)"] > media_sono["Médico(a)"]:
    print("Médicos(as) dormem menos que enfermeiros(as).")
else:
    print("Ambos dormem a mesma média de horas.")
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)


```python
# Função para separar e calcular a média da pressão sanguínea
def calcular_media_pressao(row):
    try:
        # Separar o valor de pressão sistólica e diastólica
        pressao = row.split('/')
        pressao_sistolica = float(pressao[0])  # Pressão sistólica
        pressao_diastolica = float(pressao[1])  # Pressão diastólica
        # Retornar a média das duas pressões
        return (pressao_sistolica + pressao_diastolica) / 2
    except:
        return None  # Retorna None se houver erro na conversão

# Aplicar a função para converter os valores de pressão sanguínea
df["Pressão Sanguínea"] = df["Pressão Sanguínea"].apply(calcular_media_pressao)

# Verificar se a conversão foi bem-sucedida
print(df["Pressão Sanguínea"].head())

# Agrupar os dados por Gênero e calcular a média da pressão sanguínea
media_pressao = df.groupby("Gênero")["Pressão Sanguínea"].mean()

# Exibir o resultado
print(media_pressao)

# Comparação
if media_pressao["Homem"] > media_pressao["Mulher"]:
    print("Homens têm maior pressão sanguínea média.")
elif media_pressao["Homem"] < media_pressao["Mulher"]:
    print("Mulheres têm maior pressão sanguínea média.")
else:
    print("Homens e mulheres têm a mesma pressão sanguínea média.")
```

9. É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?


```python
# Calcular a moda da coluna "Duração do sono"
moda_sono = df["Duração do sono"].mode()[0]

# Exibir o resultado
print(f"A moda da duração do sono é {moda_sono} horas.")

# Verificar se é predominante dormir 8 horas
if moda_sono == 8:
    print("É predominante entre os participantes dormir 8 horas por dia.")
else:
    print(f"Não é predominante, a moda é {moda_sono} horas.")
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# Criar os dois grupos com base na frequência cardíaca
grupo_maior_70 = df[df["Frequência cardíaca"] > 70]
grupo_menor_igual_70 = df[df["Frequência cardíaca"] <= 70]

# Calcular a média de passos para cada grupo e arredondar para inteiro
media_passos_maior_70 = round(grupo_maior_70["Passos diários"].mean())
media_passos_menor_igual_70 = round(grupo_menor_igual_70["Passos diários"].mean())

# Exibir os resultados
print(f"Média de passos (FC > 70): {media_passos_maior_70}")
print(f"Média de passos (FC <= 70): {media_passos_menor_igual_70}")

# Comparação
if media_passos_maior_70 > media_passos_menor_igual_70:
    print("Pessoas com frequência cardíaca acima de 70 dão mais passos.")
elif media_passos_maior_70 < media_passos_menor_igual_70:
    print("Pessoas com frequência cardíaca menor ou igual a 70 dão mais passos.")
else:
    print("Pessoas com frequência cardíaca acima de 70 e abaixo de 70 dão a mesma quantidade de passos.")
```
